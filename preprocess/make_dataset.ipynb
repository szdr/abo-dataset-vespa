```python
import pandas as pd
import pathlib
import numpy as np
```

```python
pd.options.display.max_colwidth = 100
```

```python
dataset_root = "../dataset"
```

```python
df_listings_list = [pd.read_json(p.absolute(), lines=True) for p in pathlib.Path(dataset_root + "/listings/metadata").glob("*")]
```

```python
df_listings = pd.concat(df_listings_list)
```

```python
df_listings.head()
```

```python
df_listings.columns
```

```python
df_listings["country"].value_counts()
```

```python
df_listings_us = df_listings.query("country == 'US'")
```

```python
df_listings_us.head()
```

```python
df_images = pd.read_csv(dataset_root + "/images/metadata/images.csv")
```

```python
df_images
```

```python
df_target = df_listings_us.drop_duplicates(
    subset=["main_image_id"]
).merge(
    df_images,
    left_on="main_image_id",
    right_on="image_id"
)
```

```python
df_target["item_name_en_us"] = df_target.item_name.apply(
    lambda arr: next(filter(lambda e: e["language_tag"] == "en_US", arr))["value"]
)
```

```python
df_target[["item_id", "item_name_en_us", "path"]]
```

```python
df_target[["item_id", "item_name_en_us", "path"]].to_csv("../output/dataset.tsv", index=False, sep="\t")
```
